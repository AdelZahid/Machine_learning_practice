```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
plt.quiver(0, 0, 4, 5, angles='xy', scale_units='xy', scale=1)
plt.show()
```

```python
plt.quiver(0, 0, -2, 3, angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(-3, 5)
plt.ylim(-1, 6)
plt.show()
```

```python
plt.quiver(0,0,4,5, angles='xy', scale_units='xy', scale=1)
plt.quiver(0,0,-2,3, angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(-3, 5)
plt.ylim(-1, 6)
plt.show()
```

```python
vector1 = np.array([0,0,4, 5])
vector2 = np.array([0,0,-2, 3])
vector_sum = vector1 + vector2
print(vector_sum)
```

```python
#addition of two vectors
plt.quiver(0,0,4,5, angles='xy', scale_units='xy', scale=1)
plt.quiver(0,0,-2,3, angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0,0,2,8, angles='xy', scale_units='xy', scale=1, color='g')
plt.xlim(-3, 5) 
plt.ylim(-1, 9)
plt.show()
```

```python
#subtraction of two vectors
vector_1 = np.array([0,0,4, 5])
vector_2 = np.array([0,0,-2, 3])
vector_sub = vector_1 - vector_2
print(vector_sub)
```

```python
#plotting the subtraction result,resultant vector is yellow
plt.quiver(0,0,4,5, angles='xy', scale_units='xy', scale=1)
plt.quiver(0,0,-2,3, angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0,0,6,2, angles='xy', scale_units='xy', scale=1, color='y')
plt.xlim(-3, 7) 
plt.ylim(-1, 6)
plt.show()
```

```python
#dot product of two vectors
vector_a = np.array([4, 5])
vector_b = np.array([-2, 3])
dot_product = np.dot(vector_a, vector_b)
print(dot_product)
```

```python
#cross product of two vectors
vector_x = np.array([4, 5, 6])
vector_y = np.array([-2, 3, 4])
cross_product = np.cross(vector_x, vector_y)
print(cross_product)
```

```python
#projection of vector_a on vector_b
vector_a = np.array([4, 5])
vector_b = np.array([-2, 3])
unit_vector_b = vector_b / np.sqrt(np.dot(vector_b, vector_b))
projection_length = np.dot(vector_a, unit_vector_b)
projection_vector = projection_length * unit_vector_b
print(projection_vector)
```
