```python
import numpy as np
import matplotlib.pyplot as plt

random_matrix = np.random.randint(100, size=(5,6))
print(random_matrix)
plt.imshow(random_matrix, cmap='viridis')
plt.colorbar()
plt.title('Random Matrix Visualization')
plt.show()
```

```python
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])
matrix_sum = matrix_a + matrix_b
print("Matrix A:\n", matrix_a)
print("Matrix B:\n", matrix_b)
print("Sum of Matrix A and B:\n", matrix_sum)
matrix_product = np.dot(matrix_a, matrix_b)
print("Product of Matrix A and B:\n", matrix_product)
```

```python
identity_matrix = np.eye(3)
print("Identity Matrix:\n", identity_matrix)
```

```python
new_mat=np.random.randint(10, size=(4,5))
transposed_mat = new_mat.transpose()
print("Original Matrix:\n", new_mat)
print("Transposed Matrix:\n", transposed_mat)
```
