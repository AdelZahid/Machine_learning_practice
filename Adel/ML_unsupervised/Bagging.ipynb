```python
import pandas as pd
import numpy as np

df=pd.read_csv('diabetes.csv')
df
```

```python
df.describe()
```

```python
X=df.drop(['Outcome'],axis='columns')
X
```

```python
Y=df['Outcome']
Y
```

```python
numerical_columns = X.select_dtypes(include=['float64', 'int64']).columns
```

```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
X_scaled=Scaler.fit_transform(X[numerical_columns])
X_scaled
```

```python

x=pd.DataFrame(X_scaled,columns=numerical_columns)
x
```

Stratify is used to devide the test and train test data devinely or balanced

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,Y,stratify=Y,random_state=20)

x_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
Score=cross_val_score(DecisionTreeClassifier(),x,Y,cv=5)
Score
```

```python
Score.mean()
```

```python
from sklearn.ensemble import BaggingClassifier

bag_model=BaggingClassifier(estimator=DecisionTreeClassifier(),n_estimators=100,max_samples=0.8,oob_score=True,random_state=0)
bag_model.fit(x_train,y_train)
bag_model.oob_score_
```
