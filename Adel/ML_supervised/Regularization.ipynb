```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df=pd.read_csv('Melbourne_housing_FULL.csv')
df.head()
```

```python
obj_column=df.select_dtypes(include=['object'])

obj_column.head()
```

```python
reduce_df=df.drop(['Address','Date','Postcode'],axis='columns')
reduce_df
```

```python
reduce_df.isnull()
```

```python
reduce_df.isna().sum()
```

here we will fill some values with zero because this values are not missing they are not given because this valuse are mena to zero such as for bedroom , bathroom,cars the missing value indicates that there is no bedroom or bathroom for that flat

```python
cols_to_zero=['Propertycount','Car','Bathroom','Bedroom2']
reduce_df[cols_to_zero]=reduce_df[cols_to_zero].fillna(0)
reduce_df
```

```python
fill_values = {}

for col in reduce_df.columns:
    if reduce_df[col].dtype == 'object':   # categorical/text columns
        fill_values[col] = reduce_df[col].mode()[0]
    else:                                  # numeric columns
        fill_values[col] = reduce_df[col].mean()

reduce_df = reduce_df.fillna(fill_values)
reduce_df
```

```python
reduce_df.isna().sum()
```

```python
dataset=pd.get_dummies(reduce_df,drop_first=True)
dataset
```

```python
X=dataset.drop('Price',axis='columns')
Y=dataset['Price']
X
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(x_train,y_train)
reg.score(x_test,y_test)
```

```python
from sklearn import linear_model

lesso_reg = linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lesso_reg.fit(x_train,y_train)
lesso_reg.score(x_test,y_test)
```
