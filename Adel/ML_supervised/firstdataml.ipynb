```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model
import sklearn.neighbors

# Load the data
dataframe = pd.read_csv("better-life-index-2024.csv", thousands=',')

# Clean column names by removing leading/trailing spaces
dataframe.columns = dataframe.columns.str.strip()

# Prepare data
X = np.c_[dataframe["GDP per capita (USD)"]]  # 2D array for training
y = np.c_[dataframe["Life expectancy"]]  # 2D array for target

# Select and train model
model = sklearn.neighbors.KNeighborsRegressor(n_neighbors=3)
model.fit(X, y)

# Get user input with validation
while True:
    try:
        user_input = input("Enter the GDP per capita of any country (USD): ").strip()
        if not user_input:  # Check for empty input
            raise ValueError("Input cannot be empty. Please enter a number.")
        x = float(user_input)  # Use float to allow decimal inputs
        if x < 0:
            raise ValueError("GDP cannot be negative. Please enter a valid number.")
        break
    except ValueError as e:
        print(f"Error: {e}. Try again.")

# Prepare input for prediction (2D array)
x_pred = np.array([[x]])  # Reshape to 2D array

# Make prediction
prediction = model.predict(x_pred)
print(f"Predicted Life expectancy for GDP per capita {x:.2f} USD: {prediction[0][0]:.2f} years")

# Plot scatter with regression line
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, model.predict(X), color='red', label='Regression line')
plt.xlabel("GDP per capita (USD)")
plt.ylabel("Life expectancy")
plt.title("GDP per capita vs Life expectancy")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split

# Load the data
dataframe = pd.read_csv("better-life-index-2024.csv", thousands=',')

# Clean column names by removing leading/trailing spaces
dataframe.columns = dataframe.columns.str.strip()

# Prepare data
X = np.c_[dataframe["GDP per capita (USD)"]]  # 2D array for features
y = dataframe["Life expectancy"]  # Continuous target

# Create binary labels based on median life expectancy
median_life_expectancy = y.median()
y_binary = (y > median_life_expectancy).astype(int)  # 1 for above median, 0 for below

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.3, random_state=42)

# Train a RandomForestClassifier
model = LogisticRegression( random_state=42)
model.fit(X_train, y_train)

# Get predicted probabilities for the positive class (needed for ROC curve)
y_scores = model.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
fpr, tpr, _ = roc_curve(y_test, y_scores)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')  # Diagonal line (random classifier)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve: GDP per capita vs Life Expectancy (Binary) - Logistic Rigression')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

# Print AUC
print(f"Area Under the ROC Curve (AUC): {roc_auc:.2f}")
```

```python
import os
import tarfile
from six.moves import urllib
import pandas as pd

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"  # Fixed typo: 'nousing' to 'housing'

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()

def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    return pd.read_csv(csv_path)

# Call fetch_housing_data to download and extract the file
fetch_housing_data()

# Now load the data
housing = load_housing_data()
housing.head()
```
