```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits

digit=load_digits()
digit.feature_names
```

```python
df=pd.DataFrame(digit.data)
df
```

```python
df['target']=digit.target
df
```

```python
X=df.drop(['target'],axis='columns')
Y=df['target']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
from sklearn.neighbors import kneighbors_graph

# Create a KNN graph with 5 neighbors, using connectivity mode
knn_graph = kneighbors_graph(X_scaled, n_neighbors=5, mode='connectivity', include_self=False)
knn_graph
```

```python
print(knn_graph.shape)  # Shape: (1797, 1797)
print(knn_graph.nnz)    # Number of non-zero entries

# View a small portion (e.g., first 5 rows) as dense
print(knn_graph[:5, :5].toarray())
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Convert a small subset of the graph to networkx
small_graph = knn_graph[:50, :50]  # First 50 samples
G = nx.from_scipy_sparse_array(small_graph)

plt.figure(figsize=(8, 8))
nx.draw(G, with_labels=True, node_size=50, font_size=8)
plt.title("KNN Graph (First 50 Samples)")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(digit.data,df['target'],test_size=0.2)
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier

model=KNeighborsClassifier(n_neighbors=10)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix

y_pred=model.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
%matplotlib inline
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(cm,annot=True)
plt.xlabel('prediction')
plt.ylabel('True')
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
